```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv(r"C:\Users\harsh\OneDrive\Documents\netflix_database.csv")

print(df.shape)
df.head()
```

```python
df.info()

df.describe(include='all')

df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='type'
)

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

```python
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)

df['year_added'] = df['date_added'].dt.year
```

```python
plt.figure(figsize=(12,5))

df['year_added'].value_counts().sort_index().plot(
    kind='line',
    marker='o'
)

plt.title("Netflix Content Added By Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
plt.figure(figsize=(12,5))

df['year_added'].value_counts().sort_index().plot(
    kind='line',
    marker='o'
)

plt.title("Netflix Content Added By Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
plt.figure(figsize=(10,5))

df['rating'].value_counts().head(10).plot(
    kind='bar'
)

plt.title("Top Ratings on Netflix")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(12,6))

top_countries.plot(
    kind='bar'
)

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Titles")

plt.show()
```

```python
content_year = pd.crosstab(
    df['release_year'],
    df['type']
)

content_year.tail(20).plot(
    figsize=(12,6)
)

plt.title("Movies and TV Shows Released Over Time")
plt.xlabel("Year")
plt.ylabel("Count")

plt.show()
```

```python
top_directors = (
    df[df['director']!='Unknown']
    ['director']
    .value_counts()
    .head(10)
)

plt.figure(figsize=(12,5))

sns.barplot(
    x=top_directors.values,
    y=top_directors.index
)

plt.title("Top 10 Directors on Netflix")

plt.show()
```

```python
heatmap_data = pd.crosstab(
    df['release_year'],
    df['type']
)

plt.figure(figsize=(12,8))

sns.heatmap(
    heatmap_data,
    cmap='YlGnBu'
)

plt.title("Release Year vs Content Type")

plt.show()
```
