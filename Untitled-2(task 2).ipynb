```python
df['GPA_Improvement'] = (
    df['Post_Semester_GPA']
    - df['Pre_Semester_GPA']
)

df[['Pre_Semester_GPA',
    'Post_Semester_GPA',
    'GPA_Improvement']].head()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Weekly_GenAI_Hours'],
    bins=30,
    kde=True
)

plt.title("Weekly GenAI Usage")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y='Primary_Use_Case',
    data=df,
    order=df['Primary_Use_Case'].value_counts().index
)

plt.title("Primary Use Cases of AI")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Burnout_Risk_Level',
    data=df
)

plt.title("Burnout Risk Levels")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='Major_Category',
    y='Weekly_GenAI_Hours',
    data=df
)

plt.xticks(rotation=45)

plt.title("AI Usage by Major")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Paid_Subscription',
    y='Post_Semester_GPA',
    data=df
)

plt.title("Paid Subscription vs GPA")
plt.show()
```

```python
numeric_cols = df.select_dtypes(
    include=np.number
)

corr_matrix = numeric_cols.corr()
plt.figure(figsize=(12,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```
